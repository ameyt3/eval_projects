```python
dat_train = ds_train.CUISINES.apply(lambda x: len(x.split(',')))
dat_test = ds_test.CUISINES.apply(lambda x: len(x.split(',')))
```

```python
print('max number of CUISINES hold by restaurant in train is ',max(dat_train))
print(ds_train.CUISINES[dat_train.idxmax()])
print('minimum number of CUISINES hold by restaurant in train is ',min(dat_train))
print(ds_train.CUISINES[dat_train.idxmin()])
```

```python
print('max number of CUISINES hold by restaurant in test is',max(dat_test))
print(ds_test.CUISINES[dat_test.idxmax()])
print('min number of CUISINES hold by restaurant in test is',min(dat_test))
print(ds_test.CUISINES[dat_test.idxmin()])
```
